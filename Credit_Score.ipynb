```python
df['CAT_GAMBLING'] = df['CAT_GAMBLING'].replace({('No',): 0, ('High',): 1, ('Low',): 2})
```

```python
df.info()
```

```python
x = df.drop('DEFAULT', axis = 1)
y = df['DEFAULT']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=42)
```

```python
rand_forst = RandomForestClassifier(random_state=12)
rand_forst.fit(x_train, y_train)
```

```python
y_pred_train = rand_forst.predict(x_train)
conf_matrix = confusion_matrix(y_pred_train, y_train)
print(conf_matrix)
prec_score = precision_score(y_pred_train, y_train)
print(prec_score)
reca_score = recall_score(y_pred_train, y_train)
print(reca_score)
accu_score = accuracy_score(y_pred_train, y_train)
print(accu_score)
```

```python
y_pred_test = rand_forst.predict(x_test)
conf_matrix = confusion_matrix(y_pred_test, y_test)
print(conf_matrix)
prec_score = precision_score(y_pred_test, y_test)
print(prec_score)
reca_score = recall_score(y_pred_test, y_test)
print(reca_score)
accu_score = accuracy_score(y_pred_test, y_test)
print(accu_score)
```
